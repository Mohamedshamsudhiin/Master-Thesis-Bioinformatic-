# General code  Euclidean Distance Between Two Columns and Add a New Column 
This code from the document "General code to extract features from DepMap"

```python
# General function to calculate Euclidean distance
def add_euclidean_distance(df, col1, col2, new_col_name):
    """
    Calculate the Euclidean distance between two columns and add it as a new column.
    The function relay on pandas and numpy 
    
    Parameters:
    df (pd.DataFrame): The DataFrame where the calculation is performed.
    col1 (str): The name of the first column (for distance calculation).
    col2 (str): The name of the second column (for distance calculation).
    new_col_name (str): The name of the new column to store the Euclidean distances.
    
    Returns:
    pd.DataFrame: The DataFrame with the new Euclidean distance column added.
    """
    # Calculate Euclidean distance for each row and add it as a new column
    df.loc[:, new_col_name] = np.sqrt((df[col1] - df[col2]) ** 2)
    return df
```

# Create a new calass_label that assign for commun GI and no 


```python
# Define a function to assign "GI" or "no GI"
def assign_class_label_neg(row):
    labels = [row["Class_label_sum"], row["Class_label_dep"], row["Class_label_corr"], row["Class_label_exp"]]
    if all(label == "no GI" for label in labels):
        return "no GI"
    elif all(label == "GI" for label in labels):
        return "GI"
    else:
        return "Mixed"  

# Create the new column
df_feature_label["Class_label_neg"] = df_feature_label.apply(assign_class_label_neg, axis=1)
```

# General code to cluster columns 

```python

def plot_correlation_matrix(df, include_columns=None, title="Correlation Matrix"):
    """
    This function calculates and visualizes the correlation matrix for numeric columns in the DataFrame.
    It also generates a heatmap and pairplot for visual analysis.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    include_columns (list, optional): List of columns to include for correlation analysis. If None, all numeric columns will be considered.
    title (str): Title for the plots. Default is 'Correlation Matrix'.

    Returns:
    None: Displays the correlation matrix heatmap and pairplot.
    """
    try:
        # If include_columns is provided, select only those columns, else use all numeric columns
        if include_columns:
            df = df[include_columns]
        
        # Select numeric columns (either int or float)
        numeric_columns = df.select_dtypes(include=['number']).columns

        # Display the numeric columns
        print(f"Numeric columns: {numeric_columns}")

        # Compute the correlation matrix
        correlation_matrix = df[numeric_columns].corr().abs()

        # Plot correlation matrix as a heatmap
        plt.figure(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
        plt.title(f"{title} - Heatmap")
        plt.show()

        # Plot pairplot (scatterplot matrix) for numeric columns
        sns.pairplot(df[numeric_columns], kind="scatter")
        plt.suptitle(f"{title} - Pairplot", y=1.02)  
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

```
